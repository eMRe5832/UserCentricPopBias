```python
import numpy as np; np.random.seed(0)
import seaborn as sb
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import matplotlib as mpl
from matplotlib.cm import ScalarMappable
```

```python
aplt = pd.read_csv("Dianping/APLT-Dianping.csv", delimiter=';')
f1 = pd.read_csv("Dianping/F1-Dianping.csv", delimiter=';')
ndcg = pd.read_csv("Dianping/NDCG-Dianping.csv", delimiter=';')
novelty = pd.read_csv("Dianping/Novelty-Dianping.csv", delimiter=';')
apltTest = pd.read_csv("Dianping/APLTTest-Dianping.csv", delimiter=';')
f1Test = pd.read_csv("Dianping/F1Test-Dianping.csv", delimiter=';')
ndcgTest = pd.read_csv("Dianping/NDCGTest-Dianping.csv", delimiter=';')
noveltyTest = pd.read_csv("Dianping/NoveltyTest-Dianping.csv", delimiter=';')
```

```python
# nDCG için önce size;
data = [[], []]
labels = [[], ['', '', '', '', '', '', '', '', '', '']]
data[0]=ndcg.iloc[0].tolist()
data[1]=ndcg.iloc[1].tolist()
labels[0]=ndcgTest.iloc[0].tolist()

ax = sb.heatmap(data, annot=labels, fmt = '', cmap="Blues")
ax.set_xticklabels(['Random', 'ItemAVG', 'MMMF', 'WMF', 'HPF', 'IBPR', 'WBPR', 'SKM', 'NEUMF', 'VAECF'])
ax.set_yticklabels(['High', 'Mid'])
ax.tick_params(axis='x', rotation=280)
ax.tick_params(axis='y', rotation=0)
plt.savefig(f'Figures/Dianping/nDCG_S_Dianping.pdf', dpi=300, bbox_inches='tight')
```

```python
# ndcg with loop for all groups
i=3
k=2
for group in ['M', 'MD','ME','DE','MED']:
    data = [[], [], []]
    labels = [[], [], ['', '', '', '', '', '', '', '', '', '']]
    data[0]=ndcg.iloc[i].tolist()
    data[1]=ndcg.iloc[i+1].tolist()
    data[2]=ndcg.iloc[i+2].tolist()
    labels[0]=ndcgTest.iloc[k].tolist()
    labels[1]=ndcgTest.iloc[k+1].tolist()
    i=i+4
    k=k+3
    fig, ax = plt.subplots()
    ax = sb.heatmap(data, annot=labels, fmt = '', cmap="Blues")
    ax.set_xticklabels(['Random', 'ItemAVG', 'MMMF', 'WMF', 'HPF', 'IBPR', 'WBPR', 'SKM', 'NEUMF', 'VAECF'])
    ax.set_yticklabels(['High', 'Mid', 'Low'])
    ax.tick_params(axis='x', rotation=280)
    ax.tick_params(axis='y', rotation=0)
    plt.savefig(f'Figures/Dianping/nDCG_{group}_Dianping.pdf', dpi=300, bbox_inches='tight')
```
