```python
import numpy as np; np.random.seed(0)
import seaborn as sb
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import matplotlib as mpl
from matplotlib.cm import ScalarMappable
sb.set(style="darkgrid")
```

```python
gap = pd.read_csv("DB/GAP-DB.csv", delimiter=';')
algo_names = ['Random', 'ItemAVG', 'MMMF', 'WMF', 'HPF', 'IBPR', 'WBPR', 'SKM', 'NEUMF', 'VAECF']

```

```python
# GAP figure for size 

medium_gap_vals = []
high_gap_vals = []

for i in range(0, len(algo_names)):
    high_gap_vals=gap.iloc[0].tolist()
    medium_gap_vals=gap.iloc[1].tolist()

barWidth = 0.07
bars1 = [high_gap_vals[0], medium_gap_vals[0]]
bars2 = [high_gap_vals[1], medium_gap_vals[1]]
bars3 = [high_gap_vals[2], medium_gap_vals[2]]
bars4 = [high_gap_vals[3], medium_gap_vals[3]]
bars5 = [high_gap_vals[4], medium_gap_vals[4]]
bars6 = [high_gap_vals[5], medium_gap_vals[5]]
bars7 = [high_gap_vals[6], medium_gap_vals[6]]
bars8 = [high_gap_vals[7], medium_gap_vals[7]]
bars9 = [high_gap_vals[8], medium_gap_vals[8]]
bars10 = [high_gap_vals[9], medium_gap_vals[9]]

r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]
r6 = [x + barWidth for x in r5]
r7 = [x + barWidth for x in r6]
r8 = [x + barWidth for x in r7]
r9 = [x + barWidth for x in r8]
r10 = [x + barWidth for x in r9]

patterns = [ "/" , "\\" , "|" , "-" , "+" , "x", "o", "O", ".", "*"]
# Make the plot
plt.bar(r1, bars1, width=barWidth, label='Random', hatch=patterns[0])
plt.bar(r2, bars2, width=barWidth, label='ItemAVG', hatch=patterns[1])
plt.bar(r3, bars3, width=barWidth, label='MMMF', hatch=patterns[2])
plt.bar(r4, bars4, width=barWidth, label='WMF', hatch=patterns[3])
plt.bar(r5, bars5, width=barWidth, label='HPF', hatch=patterns[4])
plt.bar(r6, bars6, width=barWidth, label='IBPR', hatch=patterns[5])
plt.bar(r7, bars7, width=barWidth, label='WBPR', hatch=patterns[6])
plt.bar(r8, bars8, width=barWidth, label='SKM', hatch=patterns[7])
plt.bar(r9, bars9, width=barWidth, label='NEUMF', hatch=patterns[8])
plt.bar(r10, bars10, width=barWidth, label='VAECF', hatch=patterns[9])

# Add xticks on the middle of the group bars + show legend
plt.xlabel('User group', fontsize='14')
plt.ylabel('% $\Delta$ GAP', fontsize='14')
plt.xticks([r + barWidth + 0.3 for r in range(len(bars1))], ['High','Mid'], fontsize='13')
plt.yticks(fontsize='10')
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0., framealpha=1, fontsize='12')
plt.savefig('Figures/DB/GAP_S_DB.pdf', dpi=300, bbox_inches='tight')
```
