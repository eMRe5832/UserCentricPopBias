```python
import numpy as np; np.random.seed(0)
import seaborn as sb
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import matplotlib as mpl
from matplotlib.cm import ScalarMappable
sb.set(style="darkgrid")
```

```python
gap = pd.read_csv("Debiasing/GAP-Imp.csv", delimiter=';')
algo_names = ['Aug', 'Mul', 'xQuad']

# Change .csv file according to desired metric 
```

```python
# GAP figure for size 

medium_gap_vals = []
high_gap_vals = []

for i in range(0, len(algo_names)):
    high_gap_vals=gap.iloc[0].tolist()
    medium_gap_vals=gap.iloc[1].tolist()

barWidth = 0.07
bars1 = [high_gap_vals[0], medium_gap_vals[0]]
bars2 = [high_gap_vals[1], medium_gap_vals[1]]
bars3 = [high_gap_vals[2], medium_gap_vals[2]]

r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]


patterns = [ "/" , "\\" , "|" ]
# Make the plot
plt.bar(r1, bars1, width=barWidth, label='Aug', hatch=patterns[0])
plt.bar(r2, bars2, width=barWidth, label='Mul', hatch=patterns[1])
plt.bar(r3, bars3, width=barWidth, label='xQuad', hatch=patterns[2])

# Add xticks on the middle of the group bars + show legend
plt.xlabel('S group', fontsize='14')
plt.ylabel('% $\Delta$ GAP', fontsize='14')
plt.xticks([r + barWidth + 0.3 for r in range(len(bars1))], ['High','Mid'], fontsize='13')
plt.yticks(fontsize='10')
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0., framealpha=1, fontsize='12')
```
